Student's solution

````python
# Write your code

class BankAccount:
    def __init__(self, his_name, firstmoney, comingmoney, goingmoney):
        self.his_name = his_name
        self.firstmoney = firstmoney
        self.comingmoney = comingmoney
        self.goingmoney = goingmoney
    def name(self):
        self.his_name
    def check_balance(self):
        self.firstmoney + self.comingmoney - self.goingmoney
    def deposit(self):
        self.comingmoney
    def withdraw(self):
        self.goingmoney
````
